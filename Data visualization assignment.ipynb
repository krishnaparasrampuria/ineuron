```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris.data.shape
```

```python
iris
```

```python
iris.target_names
```

```python
pca = decomposition.PCA(n_components=3)
X = iris.data
X = pca.fit_transform(X)
Y = iris.target
```

```python
import pandas as pd
mapping_dict = {0:'setosa',1:'versicolor',2:'virginica'}
X = pd.DataFrame(X)
Y = pd.DataFrame(Y)
df = pd.concat([X,Y],axis=1)
df.columns = ['x','y','z','A']
df['B'] = df['A'].map(mapping_dict)
df.head()
```

```python
fig = plt.figure(figsize=(6,5))
ax = Axes3D(fig, rect=[0,0,1,1])
labels = df['B'].unique()
ax.scatter(df['x'], df['y'], df['z'], s=100, c=df['A'])
plt.show()
```
